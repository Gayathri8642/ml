```python
x=wq.iloc[:,:9].values
print(x)
y=wq.iloc[:,9].values
print(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
tq=MinMaxScaler()
x=tq.fit_transform(x)
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("x_train")
print(x_train)
print("x_test")
print(x_test)
print("y_train")
print(y_train)
print("y_test")
print(y_test)
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
obj=RandomForestClassifier()
sele=RFE(obj)
features_imp=sele.fit(x,y)
```

```python
print(features_imp.support_)
```

```python
ran=obj.fit(x_train,y_train)
```

```python
y_pred=ran.predict(x_test)
print(y_pred)
```

```python
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
def calculate_precision_recall(y_test, y_pred):


    tp = np.sum(np.logical_and(y_test == 1, y_pred == 1))
    fn = np.sum(np.logical_and(y_test == 1, y_pred == 0))
    fp = np.sum(np.logical_and(y_test == 0, y_pred == 1))

    precision = tp / (tp + fp)
    recall = tp / (tp + fn)

    return precision, recall
```

```python
precision, recall = calculate_precision_recall(y_test, y_pred)
print("Precision:", precision*100)
print("Recall:", recall*100)
```
